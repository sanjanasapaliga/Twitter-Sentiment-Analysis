```python
import re
import tweepy
from tweepy import OAuthHandler
from textblob import TextBlob

class TwitterClient(object):
    def __init__(self):
        consumer_key='LlnXUxe64nBJQs0vCFPTj6Vkx'
        consumer_secret='PyhmahYOAONBaDBdw5U2mnIXCw6XdGKza0kgw35Dl86du5EegD'
        access_token='4313447914-5qh3YuUbTegPvjTMhfmeQONudsQgBDVJtHuijHB'
        access_token_secret='DBizwMGRuSemXtq3HbOaatBHa7rFXUywofhwYQY91Zj9N'
        
        self.auth=OAuthHandler(consumer_key,consumer_secret)
        self.auth.set_access_token(access_token,access_token_secret)
        self.api=tweepy.API(self.auth)
    def cleaned_tweets(self,tweet):
        return ' '.join(re.sub('',' ',tweet).split())
    def tweet_sentiments(self,text):
        analysis=TextBlob(self.cleaned_tweets(text))
        if analysis.sentiment.polarity>0:
            return 'Positive'
        else :
            return 'Negative'
    def get_tweets(self,query,count=10):
        tweets=[]
        fetch_tweets=self.api.search_tweets(q=query,count=count)
        for tweet in fetch_tweets:
            parsed_tweets={}
            parsed_tweets['text']=tweet.text
            parsed_tweets['sentiment']=self.tweet_sentiments(tweet.text)
            if tweet.retweet_count>0:
                if tweet not in tweets:
                    tweets.append(parsed_tweets)
            else:
                tweets.append(parsed_tweets)
                
        return tweets
def main():
    api=TwitterClient()
    tweets=api.get_tweets(query='modi',count=50)
    ptweets=[tweet for tweet in tweets if tweet['sentiment']=='Positive']
    print('% of positive tweets : {}%'.format(100*len(ptweets)/len(tweets)))
    ntweets=[tweet for tweet in tweets if tweet['sentiment']=='Negative']
    print('% of Negative tweets : {}%'.format(100*len(ntweets)/len(tweets)))
    
    print('/n positive tweets')
    for tweet in ptweets:
        print(tweet['text'])
    for tweet in ntweets:
        print(tweet['text'])
        
if __name__=='__main__':
    main()
```
